```python
class Celular:
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} se ha encendido.")

    def mostrar_info(self):
        print(f"Celular: {self.marca} {self.modelo}, Precio: ${self.precio}")
# Creacion de objetos(intancias)
celular1 = Celular("Samsung", "Galaxy S23", 800)
celular2 = Celular("Apple", "iPhone 13", 1000)

#Llamadas a las funcionespara cada objeto celular 1 y celular 2
print("Celular 1:")
celular1.mostrar_info()
celular1.encender()
print("------------------------------------------------")
print("\nCelular 2:")
celular2.mostrar_info()
celular2.encender()

print("------------------------------------------------")
print("Prpgrama terminado")
print(" Creador de software por Iris Chavez NL = 11")
```
